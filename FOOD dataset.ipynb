# Food Balancesheet Data-cleaning & Data Wrangling

```python
import pandas as pd
```

```python
#we the missing values with mean because more than 5% of the data is missing
FD['Y2014'].fillna(FD['Y2014'].mean(),
                  inplace=True)

FD['Y2015'].fillna(FD['Y2015'].mean(),
                  inplace=True)

FD['Y2016'].fillna(FD['Y2016'].mean(),
                  inplace=True)

FD['Y2017'].fillna(FD['Y2017'].mean(),
                  inplace=True)

FD['Y2018'].fillna(FD['Y2018'].mean(),
                  inplace=True)

FD.isna().sum()
```

```python
FD
```

```python
FD.dropna(inplace=True)#i droped the unit error cause its not significant in our data set
```
